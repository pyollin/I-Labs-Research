Here I take another approach in order to increase the SNR. Signal
here is referring to the generated frequency. 

```python
#   info of functions found in FunFile in github page:
#   https://github.com/pyollin/I-Labs-Research/blob/master/Funfile.py

import Funfile as Fun
```

```python
#   params dictionary to be used to hold import values
params = {
    'f_ext': 'Hz_10v_erm_raw.fif',
    'rec': {},
    '5': {'start': 80, 'stop': 120},
    '20': {'start': 0, 'stop': 40},
    '70': {'start': 70, 'stop': 110},
    '105': {'start': 65, 'stop': 105}
}

#   empty dictionary to hold RecClass objects
dat = {}

recs = ['5', '20', '70', '105']

for i in recs:
    params['rec'] = i
    dat[i] = Fun.RecClass(params['rec']+params['f_ext'],
                       params[i]['start'],
                       params[i]['stop'])
del i
```

Plotting raw signal vector norm vs the signal vector norm after averaging.

Plots are as
follows: 5Hz, 20Hz, 70Hz, then 105Hz. Magnetometer signal on top and
gradiometer on bottom.

```python
for i in recs:
    Fun.plot_ave_vs_raw(dat[i].data, dat[i].raw.info)
del i
```

Applying SSS to averaged signal. Note the raw signal is the the raw signal
vector norms after averaging. Filtered using SSS and using fine calibration file.

Plots are as
follows: 5Hz, 20Hz, 70Hz, then 105Hz. Magnetometer signal on top and
gradiometer on bottom.

```python
for i in recs:
    Fun.plot_ave_norms(dat[i].data, dat[i].raw.info)
del i
```

Calculating the drops of the averaged signal

```python
drops = {}

for i in recs:
    drops[i] = Fun.ave_drop(dat[i].data, dat[i].raw.info)
del i


#   creates a header for printing the table of values from drop
print("{:<8} {:<15} {:<15}".format('Hz',
                                                 'mag drop',
                                                 'grad drop',))

#   for looping through all recording and printing calculated drops
#   currently rounded to 4th decimal place, change as desired
for i in ['5','20','70','105']:
    print("{:<8} {:<15} {:<15}".format(i,
                                                     round(drops[i]['mcdat'], 4),
                                                     round(drops[i]['gcdat'], 4)))
#   deleting i from the for loop
del i
```
